```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from the two files
path1 = 'kcity_final_parking_lot3_forward_fixed.txt'
path2 = 'kcity_final_parking_lot3_reverse_fixed.txt'

# Read the data
data1 = pd.read_csv(path1, delim_whitespace=True, header=None, names=['X', 'Y', 'Z'])
data2 = pd.read_csv(path2, delim_whitespace=True, header=None, names=['X', 'Y', 'Z'])

# Ensure both paths have the same length by truncating the longer path
min_length = min(len(data1), len(data2))
data1 = data1.iloc[:min_length]
data2 = data2.iloc[:min_length]

# Define a weight for the first path
weight = 0.6

# Calculate the intermediate path, closer to the first path
mid_path = pd.DataFrame({
    'X': weight * data1['X'] + (1 - weight) * data2['X'],
    'Y': weight * data1['Y'] + (1 - weight) * data2['Y'],
    'Z': weight * data1['Z'] + (1 - weight) * data2['Z']
})

# Save the intermediate path to a txt file
# mid_path.to_csv('intermediate_path_output.txt', sep='\t', header=False, index=False)

# Plot the paths
plt.figure(figsize=(10, 8))
plt.plot(data1['X'], data1['Y'], label='Path 1', linestyle='--')
plt.plot(data2['X'], data2['Y'], label='Path 2', linestyle='--')
plt.plot(mid_path['X'], mid_path['Y'], label='Intermediate Path (Closer to Path 1)', color='purple')

# Add titles and labels
plt.title('Intermediate Path between Two Paths (Closer to Path 1)')
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.legend()
plt.grid(True)
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from the two files
path1 = 'kcity_final_parking_lot2_forward_fixed.txt'
path2 = 'kcity_final_parking_lot2_return_fixed.txt'
path3 = 'kcity_final_parking_lot2_reverse_fixed.txt'

# Read the data
data1 = pd.read_csv(path1, delim_whitespace=True, header=None, names=['X', 'Y', 'Z'])
data2 = pd.read_csv(path2, delim_whitespace=True, header=None, names=['X', 'Y', 'Z'])
data3 = pd.read_csv(path3, delim_whitespace=True, header=None, names=['X', 'Y', 'Z'])

# Ensure both paths have the same length by truncating the longer path
min_length = min(len(data1), len(data2))
data1 = data1.iloc[:min_length]
data2 = data2.iloc[:min_length]
data2 = data3.iloc[:min_length]


# Plot the paths
plt.figure(figsize=(10, 8))
plt.plot(data1['X'], data1['Y'], label='Path 1', linestyle='--')
plt.plot(data2['X'], data2['Y'], label='Path 2', linestyle='--')
plt.plot(data3['X'], data3['Y'], label='Path 3', linestyle='--')

# Add titles and labels
plt.title('Intermediate Path between Two Paths (Closer to Path 1)')
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.legend()
plt.grid(True)
plt.show()
```
